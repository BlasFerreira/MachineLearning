# Import

## Library

```python
#  Extraction Information 
import requests
import lxml.html as html
import time 
import json
import numpy as np

# Librarys for count word
from collections import Counter
from traceback import print_tb
import re
import xlwt
from xlwt import Workbook
import pandas
```

# Information Extraction 

## Constants

```python
URL = 'https://www.ted.com/talks?page='
LINK_TO_ARTICLE = '//div[@class="media__image media__image--thumb talk-link__image"]//@href'
LINK_TO_BODY = '//script[contains(@id,"__NEXT_DATA__")]'
# Amoun ted talk for extraccion her transcript. For example just one page
TEDS = 1 + ( 1 )
```

```python
FILE_ORIGIN = 'ExtraccionDeInformacion.txt'
# Word more frecuntly that count
NUMBER_OF_WORDS = 3000
```

## Algorithm for transcript extraccion 

```python
# Gerate link secundary page=1,page=2, etc.
def generador_de_url( url, teds ):
    urls_generado = [] # Create a array for gather link
    for i in range( 1,teds ) :
        urls_generado.append( url + str(i) )    
    return urls_generado


def extraccion_links_page( urls ):
    try:
        response = requests.get( urls )
        if response.status_code == 200 :

            response2 = response.content.decode('utf-8')
            response3 = html.fromstring(response2)
            link_articulos = response3.xpath(LINK_TO_ARTICLE)

            return link_articulos
        else:
            raise ValueError(f'Error : {response.status_code}')
    except ValueError as ve:
        print(ve)


def extraccion_de_body(link_talk,link_body):
    try:

        r1 = requests.get( link_talk )
        rr = r1.content
        if r1.status_code == 200 :  
            r2 = rr.decode('utf-8')
            r3 = html.fromstring(r2)
            body = r3.xpath( link_body )            
            open('facu3.html','wb').write(rr)            
            jsondata = json.loads(body[0].text.strip()) 
            jsondata2 = json.loads(jsondata["props"]["pageProps"]["videoData"]["playerData"]) 
            rTra = jsondata2["resources"]["hls"]["metadata"]            
            rrr = requests.get(rTra)
            jsondata3 = json.loads(rrr.content) 
            i=0
            try:
                # keep language of the transcript
                language = jsondata3["subtitles"][i]["name"]
                if language == 'English': 
                
                    # rTra2 keep link to transcript
                    rTra2 = jsondata3["subtitles"][i]["webvtt"]                
                    rrrr = requests.get(rTra2)
                    rTra3 = rrrr.content
                    rTra3 = rTra3.decode('utf-8')
                    rTra4 = rTra3.replace('WEBVTT','').replace('?','').replace(',','').replace('>','').replace('-','').replace('.','').replace('0','').replace('1','').replace('2','').replace('3','').replace('4','').replace('5','').replace('6','').replace('7','').replace('8','').replace('9','').replace(':','')
#                     print(rTra4)
                    f = open("ExtraccionDeInformacion.txt","a")
                    f.write(rTra4)
                    f.close()
                
                else :

                    while language != 'English' :
#                         print(f'    The language is {language}')
                        i+=1
                        language = jsondata3["subtitles"][i]["name"]
#                         print(f'    The language is {language}')
                        time.sleep(2)
                        if language == 'English': 
                        
                            # rTra2 keep link to transcript
                            rTra2 = jsondata3["subtitles"][i]["webvtt"]           
                            rrrr = requests.get(rTra2)
                            rTra3 = rrrr.content
                            rTra3 = rTra3.decode('utf-8')
                            rTra4 = rTra3.replace('WEBVTT','').replace('?','').replace(',','').replace('>','').replace('-','').replace('.','').replace('0','').replace('1','').replace('2','').replace('3','').replace('4','').replace('5','').replace('6','').replace('7','').replace('8','').replace('9','').replace(':','')
#                             print(rTra4)
                            f = open("ExtraccionDeInformacion.txt","a")
                            f.write(rTra4)
                            f.close()
            except IndexError :
                print(' NO TRANSCRIPT')

        else:
            raise ValueError(f'Error : {r1.status_code}' )
    except ValueError as ve:
        print(ve)



if __name__=='__main__':
    all_urls = []
    i=0
    urls_generados = generador_de_url( URL, TEDS )


    for l in urls_generados:
        print(f' Extraction links from urls_generados : {l}')
        aux = extraccion_links_page(l)
        for n in aux : 
            all_urls.append( f'https://www.ted.com{n}/transcript' )
        time.sleep(3)    


    for j in all_urls:
#         print(f'[{i+1}] Waiting for extraccion :  {j}')
        i+=1
        time.sleep(3)
        extraccion_de_body(j,LINK_TO_BODY)
#         print('\n')

    print('\n\nFINISH\n\n')
  
```

## Algorithm for transcript Extraction

```python
if __name__ =='__main__':

    words = re.findall(r'\w+', open(FILE_ORIGIN).read().lower())
    p = Counter(words).most_common(NUMBER_OF_WORDS)

    wb = Workbook()
    
    # add_sheet is used to create sheet.
    sheet1 = wb.add_sheet('Sheet 1')
    
    for i in range(0,NUMBER_OF_WORDS) :
        sheet1.write(i, 0, p[i][0] )
        sheet1.write(i, 1, p[i][1] )
        # print(p[i])
    
    wb.save('w.xls')
```
